```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python


# Create a new directed graph G
G = nx.DiGraph()

# Add edges (implicitely adds nodes)
G.add_edge('A', 'B')
G.add_edge('A', 'C')
G.add_edge('B', 'D')
G.add_edge('B', 'E')
G.add_edge('C', 'F')
G.add_edge('C', 'G')

# Draw the graph
nx.draw(G, with_labels=True)
plt.show()
```

```python
# Create a new undirected graph G
G = nx.Graph()

# Add edges (implicitely adds nodes)
G.add_edge('A', 'B')
G.add_edge('A', 'C')
G.add_edge('B', 'D')
G.add_edge('B', 'E')
G.add_edge('C', 'F')
G.add_edge('C', 'G')

# Draw the graph
nx.draw(G, with_labels=True)
plt.show()
```

```python
#Esempio di creazione grafo

G = nx.Graph()

nodes_list = [(1, {'color': 'blue'}), (2, {'color': 'red'}), (3, {'color': 'green'})]

G.add_nodes_from(nodes_list)


nx.draw(G)
```
