```python
import networkx as nx
import utility_functions as uf
import matplotlib.pyplot as plt
```

```python
# Create a new directed graph G
G = nx.Graph()

# Add edges (implicitely adds nodes)
G.add_edge('A', 'B')
G.add_edge('A', 'C')
G.add_edge('B', 'D')
G.add_edge('B', 'E')
G.add_edge('C', 'F')
G.add_edge('C', 'G')

# Draw the graph
nx.draw(G)
nodi = G.neighbors('A')
print(list(nodi))
```

```python
# Create a new undirected graph G
G = nx.Graph()

# Add edges (implicitely adds nodes)
G.add_edge('A', 'B')
G.add_edge('A', 'C')
G.add_edge('B', 'D')
G.add_edge('B', 'E')
G.add_edge('C', 'F')
G.add_edge('C', 'G')

# Draw the graph
nx.draw(G, with_labels=True)
plt.show()
```

```python
''' 
PROVE DI FUNZIONI DI LIBRERIA:
    - add_edges_from()
    - add_edge()
    - add_node()
    - add_nodes_from()
'''
G = nx.Graph()

G.add_edges_from([(1, 3), (1,2), (3, 'c')])
G.add_node(1, time='5pm', location='home')
G.add_edge(1, 2)
G.add_node("spam")        # adds node "spam"
G.add_nodes_from("ciaone")  # adds 4 nodes: 's', 'p', 'a', 'm'
G.add_edge(3, 'm')

nx.draw(G, with_labels=True)

print("nodes: ", G.nodes())
print("edges: ", G.edges())
```
